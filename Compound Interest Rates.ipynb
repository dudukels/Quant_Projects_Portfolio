```python
def cmpInt(P, r, m, t):
    return P*(1 + r/m)**(t*m) # PV to FV
    
r = 0.02 # interest rate per year OR annual interest rate; units= [interest / year]
          # r = i*m; units= [interest / no. period per year]*[no. period per year] = [interest / year]
P = 100000 # principal
m = 12 # number of interest payments per year
i = r/m # interest rate per no. of period per year OR periodical interest rate; units= [interest / no. period per year]

t = 50 # number of years

V = cmpInt(P, r, m, t)


print(f'Principal: {P}, Final Value: {round(V,2)}, FV-P: {round(V-P,2)} = {round( 100*(V-P)/P ,2)}%')
```

```python
def simpleInt(P, r, m, t):
    return P*(1+r*t) #PV to FV


P = 100000 # principal
m = 12 # number of interest payments per year
r = 0.02 # interest rate per year OR annual interest rate; units= [interest / year]
          # r = i*m; units= [interest / no. period per year]*[no. period per year] = [interest / year]
i = r/m # interest rate per no. of period per year OR periodical interest rate; units= [interest / no. period per year]

t = 50 # number of years

V = simpleInt(P, r, m, t)

print(f'Principal: {P}, Final Value: {round(V,2)}, FV-P: {round(V-P,2)} = {round( 100*(V-P)/P ,2)}%')
```

```python
#APR CALCULATION

P = 100 #principal
b = P #balance
i = 0.01 # periodic rate (monthly rate)
accumulated_int = 0 #accumulated interest

for x in range (1,13):
    bxi = b*i # dollar interest per period (month)
    
    accumulated_int = accumulated_int + bxi 
    
    print(f'month {x} : balance=${b} , interest=${bxi}')
    
FV = b + accumulated_int  # P*i*12 + P this is simple interest formula
print(f'Ending Balance after 1 year: ${FV}')
print(f'effective rate or APR = {P*(FV-P)/P}%')
    
```

```python
#APR CALCULATION - interest daily

P = 100 #principal
b = P #balance
i = 0.000329 # periodic rate (daily rate) = 12%/365
accumulated_int = 0 #accumulated interest

for x in range (1,366):
    bxi = b*i # dollar interest per period (month)
    
    accumulated_int = accumulated_int + bxi 
    
    print(f'day {x} : balance=${b} , interest=${bxi}')
    
FV = b + accumulated_int  # P*i*12 + P this is simple interest formula
print(f'Ending Balance after 1 year: ${FV}')
print(f'effective rate or APR = {P*(FV-P)/P}%')
    
```

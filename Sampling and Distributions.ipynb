Study the methods in generating samples (random vectors) or stochastic processes

Goal is how to obtain samples from these distributions

Note: you will not be able to obtain the shape of the sampling distribution with small samples i.e. n < 100

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random samples from a uniform distribution
low_value = 0 #a
high_value = 10 #b
num_samples = 100000 # each sample is within [a,b]
samples = np.random.uniform(low=low_value, high=high_value, size=num_samples)

print(f'shape of sample array = {samples.shape}')
print(f' 1/(b-a) = {1/(high_value-low_value)}')

# Display the histogram of the samples
# bins= number of intervals to divide [a,b]
# density= False=frequency, True=probability density
plt.hist(samples, bins=10, density=True, alpha=0.6, color='b') 
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Uniform Distribution")
plt.grid(True)
plt.show()
```

```python
np.round(samples)
```

Bernoulli Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Define the probability of success (e.g., heads)
p = 0.7

# Create a Bernoulli distribution object
rv = bernoulli(p)

#Note that bernoulli Random Variable can only take two numerical values (0 & 1 typically) to denote success anf failure

# Calculate the first four moments (mean, variance, skewness, kurtosis)
mean, var, skew, kurt = rv.stats(moments='mvsk')

# Display the probability mass function (PMF)
x = np.array([0, 1])
plt.stem(x, rv.pmf(x), basefmt="b")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.title("Bernoulli Distribution (p = 0.7)")
plt.xticks(x, ["Failure (0)", "Success (1)"])
plt.grid(True)
plt.show()

print(f"Mean: {mean:.2f}, Variance: {var:.2f}, Skewness: {skew:.2f}, Kurtosis: {kurt:.2f}")
```

Binomial Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 100  # Number of trials
p = 0.4  # Probability of heads

# Calculate the PMF for k = 7
k = 7
prob_7_heads = binom.pmf(k, n, p)

print(f"Probability of getting exactly 7 heads: {prob_7_heads:.4f}")

# Plot the binomial distribution
x = np.arange(0, n + 1)
plt.stem(x, binom.pmf(x, n, p), basefmt="b")
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution (n={n}, p={p})")
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Define the number of trials (n) and the probability of success (p)
n = 100  # Number of trials
p = 0.4  # Probability of success (e.g., getting heads in a biased coin flip)

# note: the binomial Random Variable is the sum of n bernoulli Random Variables
#          i.e. binomial RV is the summation of all successes

# Generate 20 random binomial variates
samples = np.random.binomial(n=n, p=p, size=1000000) # as the sample size increases, the realized samples are centered at n*p


plt.hist(samples, bins=25, density=True, alpha=0.6, color='b') 
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Binomial Distribution")
plt.grid(True)
plt.show()

print("Random Variates:")
print(samples)
```

Poisson Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define the expected rate of events (mean) = lambda
mu = 5  # Example: Expected number of events per interval

# Generate 1000 random Poisson variates
samples = poisson.rvs(mu=mu, size=100000) #using scipy.stats

# Plot the histogram of the generated samples
plt.hist(samples, bins=20, edgecolor='black', alpha=0.7, density=True)
plt.xlabel("Number of Events")
plt.ylabel("Frequency")
plt.title(f"Poisson Distribution (mu={mu})")
plt.grid(True)
plt.show()

print(samples[0:10])
```

```python
import numpy as np

# Define the expected rate of events (lambda)
lam = 5  # Example: Expected number of events per interval

# Generate 20 random Poisson variates
samples = np.random.poisson(lam=lam, size=100000) # N(t=1) ~ Poisson(lambda*t) = Poisson(lambda)
#using np

# Plot the histogram of the generated samples
plt.hist(samples, bins=20, edgecolor='black', alpha=0.7, density=True)
plt.xlabel("Number of Events")
plt.ylabel("Frequency")
plt.title(f"Poisson Distribution (mu={mu})")
plt.grid(True)
plt.show()

print("Random Variates:")
print(samples[0:10])
```

Exponential Distribution

```python
import numpy as np

# Set the scale (inverse of rate parameter λ) and the size (number of samples)
scale = 1.0  # This is β = 1/λ
size = 1000000    # Number of samples to draw

# Generate samples
samples = np.random.exponential(scale, size)


# Plot the histogram of the generated samples
plt.hist(samples, bins=20, edgecolor='black', alpha=0.7, density=True)
plt.xlabel("Number of Events")
plt.ylabel("Frequency")
plt.title(f"Exponential Distribution (1/lambda={scale})")
plt.grid(True)
plt.show()

print(samples)
```

```python
from scipy.stats import expon

# Set the scale (inverse of rate parameter λ) and the size (number of samples)
scale = 1.0  # This is β = 1/λ
size = 1000000    # Number of samples to draw

# Generate samples
samples = expon.rvs(scale=scale, size=size)

# Plot the histogram of the generated samples
plt.hist(samples, bins=20, edgecolor='black', alpha=0.7, density=True)
plt.xlabel("Number of Events")
plt.ylabel("Frequency")
plt.title(f"Exponential Distribution (1/lambda={scale})")
plt.grid(True)
plt.show()

print(samples)
```
